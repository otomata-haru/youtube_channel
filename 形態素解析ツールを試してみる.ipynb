```python
import MeCab
import time
start = time.time()
m = MeCab.Tagger()
sample_txt = "祇園精舎ぎおんしょうじゃの鐘の声、諸行無常の響きあり。沙羅双樹さらそうじゅの花の色、盛者必衰じょうしゃひっすいの理ことわりをあらはす。おごれる人も久しからず、ただ春の夜の夢のごとし。猛たけき者も遂にはほろびぬ、ひとへに風の前の塵におなじ。"
print(m.parse(sample_txt))

elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
```

```python
import spacy
import time
start = time.time()
nlp = spacy.load("ja_ginza")
doc = nlp("祇園精舎ぎおんしょうじゃの鐘の声、諸行無常の響きあり。沙羅双樹さらそうじゅの花の色、盛者必衰じょうしゃひっすいの理ことわりをあらはす。おごれる人も久しからず、ただ春の夜の夢のごとし。猛たけき者も遂にはほろびぬ、ひとへに風の前の塵におなじ。")

for sent in doc.sents:
    for token in sent:
        print(token.i, token.orth_, token.lemma_, token.pos_, token.tag_, token.dep_, token.head.i)
elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
```
